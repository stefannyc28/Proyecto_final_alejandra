## 1. Media (Promedio)
La media es la suma de todos los valores dividida entre el n煤mero de observaciones.

**F贸rmula:**

$$
\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

```

import numpy as np
valores = [8, 6, 7, 5, 9]
media = np.mean(valores)
media
```

## 2. Mediana
La mediana es el valor que divide la distribuci贸n en dos mitades iguales (cuando los datos est谩n ordenados).

```
np.median(valores)
```

## 3. Moda
La moda es el valor que m谩s se repite en un conjunto de datos.

```

from scipy import stats
stats.mode(valores, keepdims=True).mode[0]
```

## 4. Varianza
La varianza mide la dispersi贸n promedio respecto a la media.

**F贸rmula:**

$$
\text{Var}(X) = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2
$$

```
np.var(valores)
```

## 5. Desviaci贸n est谩ndar
Es la ra铆z cuadrada de la varianza. Mide la dispersi贸n en las mismas unidades que los datos.

```
np.std(valores)
```

## 6. Rango intercuart铆lico (IQR)
Es la diferencia entre el tercer y el primer cuartil. Representa el rango del 50% central de los datos.

```
iqr = np.percentile(valores, 75) - np.percentile(valores, 25)
iqr
```
